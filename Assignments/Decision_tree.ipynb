# Libraries


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
```

# Load data set

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

# Data preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=50)
```

# Train

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

# Model testing

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy:.4f}")
```

```python
report = classification_report(y_test, y_pred, target_names=data.target_names)
print("\nReporte de clasificación:\n", report)
```

```python
plt.figure(figsize=(20,10), dpi=300)
tree.plot_tree(model, filled=True, feature_names=data.feature_names, class_names=data.target_names, rounded=True)
plt.show()
```
